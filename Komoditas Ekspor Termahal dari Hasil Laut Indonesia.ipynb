## Cari tahu rata-rata harga ekspor setiap komoditas berdasarkan

```python
#Mengelompokkan data rata-rata Volume berdasarkan tahun
avg_komoditas = df2_tr1.groupby('Komoditas')['USD'].mean() #Mencari rata-rata volume untuk dikelompokkan berdasarkan tahun
avg_komoditas
```

```python
df_update = df2_tr1.assign(Harga_perKg = df2_tr1['USD']/df2_tr1['Volume'])
df_update
```

```python
df_update.isna().sum()
```

```python
#Merubah tanda NaN menjadi angka 0
df_update2= df_update.replace(np.nan, 0, regex=True)
df_update2
```

```python
#Mengelompokkan data rata-rata Volume berdasarkan Komoditas
avg_hrg_kmdperKg = df_update2.groupby('Komoditas')['Harga_perKg'].mean()
avg_hrg_kmdperKg
```

```python
komoditas = avg_hrg_kmdperKg.index
harga = avg_hrg_kmdperKg.values
```

```python
plt.figure(figsize=(20,10))
plt.scatter(komoditas, harga)
plt.plot(komoditas, harga)
plt.xlabel('Komoditas')
plt.ylabel('Harga Ekspor/Kg')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_hrg_kmdperKg.sort_values(ascending=False)
```

Dari hasil pengolahan dataset diatas dapat diperoleh bahwa Komoditas dengan jumlah ekspor termahal dimiliki oleh :
* 1. Mutiara, dengan harga 5241.22 USD/Kg atau setara Rp.75.641.549/Kg
* 2. Ikan Hias, dengan harga 16.99 USD/Kg atau setara Rp. 245.200/Kg
* 3. Sirip Hiu, dengan harga 15.43 USD/Kg atau setara Rp. 222.686/Kg

Komoditas yang dapat dilakukan pembudidayaan adalah Mutiara dan Ikan Hias, selanjutnya dapat dilakukan pemodelah untuk memprediksi harga 2 komoditas tersebut di tahun yang akan datang
